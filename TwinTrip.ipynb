```python
# 03
def flatten(t):
    return [item for sublist in t for item in sublist]
```

```python
Info_0301_list = os.listdir('0301_img')
```

```python
@app.route("/GetDomestryImage")
def getDomestryImage():
    CategoryImage = request.args.get('Num')  
    Key = CategoryImage.split('/')[2].split('.jpg')[0]
    AllImageList = flatten(list(TOP.values()))
    FinalImages = pd.Series(AllImageList)[[Key in Image for Image in AllImageList]]
    ID = [int(Image.split('_')[4].split('.jpg')[0]) for Image in FinalImages]
    tmpDF = pd.DataFrame(df.loc[ID])
    tmpDF['FinalImages'] = FinalImages.tolist()
    tmpDF = tmpDF.reset_index()
 
    Add_Info_Title =[]
    for i in Info_0301_list:
        Add_Info_Title.append(i.split('_')[0])

    AddImageList = []
    for title in tmpDF['Title']:
        ImageList = []
        for i in range(len(Add_Info_Title)):
            if Add_Info_Title[i] == title:
                ImageList.append(Info_0301_list[i])

        AddImageList.append(",".join(ImageList))

    tmpDF['AnotherImage'] = AddImageList

    return tmpDF.to_json()
```
